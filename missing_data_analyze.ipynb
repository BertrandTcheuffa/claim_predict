```python
stat, p = chi2_contingency(pd.crosstab(df["fraud_reported"], df["police_report_available"]))[:2]
V_Cramer = np.sqrt(stat/pd.crosstab(df["fraud_reported"],df["police_report_available"]).values.sum())
print("V_Cramer: ", V_Cramer, end="\n\n")
print("p_value : ", p)
```
